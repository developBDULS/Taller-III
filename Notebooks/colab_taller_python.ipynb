**Operaciones en DataFrame**

`````python
data = {
    'Marca':pd.Series(['Iphone','Samsung','Motorola','Xiaomi']),
    'Precio':pd.Series([300000,280000,265000,250000]),
    'Cantidad':pd.Series([5,10,12,11]),
    'Vendidos':pd.Series([1,5,4,6])
}
`````

`````python
df_smartphones = pd.DataFrame(data)
df_smartphones
`````

Obtenemos detalles estadísticos con **describe()**

`````python
df_smartphones.describe()
`````

Creación de una nueva columna a partir de columnas existentes

`````python
df_smartphones['Disponibles'] = df_smartphones['Cantidad'] - df_smartphones['Vendidos']
df_smartphones
`````

Creación de nueva columna obteniendo porcentaje de una columna existente

`````python
df_smartphones['Descuento 6%'] = df_smartphones.Precio * 0.94
df_smartphones
`````

`````python
df_smartphones['Ganancias'] = df_smartphones['Descuento 6%'] * df_smartphones.Vendidos
df_smartphones
`````

**Pop(nombre_columna)** nos permite eliminar la columna deseada

`````python
df_smartphones.pop('Descuento 5%')
df_smartphones
`````

Visualizamos los datos a través de plot integrado en pandas

`````python
df_smartphones.set_index('Marca')['Ganancias'].plot(kind='pie', 
                                                    ylabel='',
                                                    autopct='%.2f%%',                                                    
                                                    figsize=(8,8),
                                                    explode=(0,0.1,0,0))
`````

#### Creación de Gráfico desde matplotlib
Comparación de datos en un gráfico de multiples medidas

Definimos los valores que utilizaremos desde el dataframe


`````python
netflix = df_peliculas[df_peliculas['Netflix']==1].Age.value_counts(normalize=True)
disney = df_peliculas[df_peliculas['Disney+']==1].Age.value_counts(normalize=True)
prime = df_peliculas[df_peliculas['Prime Video']==1].Age.value_counts(normalize=True)
hulu = df_peliculas[df_peliculas['Hulu']==1].Age.value_counts(normalize=True)
`````

Obtenemos los valores a utilizar en el eje x

`````python
labels =df_peliculas.Age.unique()
labels
`````

Calculamos la cantidad de valores en el eje x

`````python
x = np.arange(len(labels))
x
`````

`````python
width = 0.2  # ancho de las barras
fig, ax = plt.subplots()
rects1 = ax.bar(x - width, netflix, width, label='Netflix',color='#AD1E1E')
rects2 = ax.bar(x, hulu, width, label='Disney+',color='#1855A3')
rects3 = ax.bar(x + width, prime, width, label='Prime Video',color='#6DA7F1')
rects4 = ax.bar(x + (width*2), prime, width, label='Hulu',color='#49C95B')

# Personalizamos titulo, texto a los ejes x e y 
ax.set_ylabel('')
ax.set_title('% distribución de peliculas por clasificacion de edad')
ax.set_xticks(np.arange(len(x)))
ax.set_xticklabels(['S/I','all','7+','13+','16+','18+'])
ax.legend()

fig.set_size_inches(10, 6)
fig.tight_layout()
plt.show()
`````

#### Nube de palabras

`````python
from wordcloud import WordCloud, STOPWORDS
`````

**Género de peliculas mas frecuentes**  
Obtenemos los valores a graficar, en este caso de la columna *genres*

`````python
genero = df_peliculas['Genres'].values
`````

Transformamos los datos a texto

`````python
genero =','.join(map(str, genero)).replace('Sci-Fi','ScienceFiction')
genero
`````

`````python
nube_de_palabras = WordCloud(background_color='white',width=900, height=500).generate(genero)
`````

`````python
fig = plt.figure(figsize=(15,20))
plt.imshow(nube_de_palabras, interpolation="bilinear")
plt.axis("off")
plt.show()
`````
